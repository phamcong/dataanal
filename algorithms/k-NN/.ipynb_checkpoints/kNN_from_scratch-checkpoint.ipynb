```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.colors import ListedColormap
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn import neighbors, datasets
from sklearn.metrics import accuracy_score
from collections import Counter
%matplotlib inline

def optimal_K_neighbors(X_train, y_train, k_fold):
    cv_scores = [] # list of cross validation scores
    # create a list of 49 elements
    myList = list(range(1, 50)) 
    # filter only odd value (K should be odd)
    neighbors_ = [x for x in myList if x%2 != 0]
    
    for k in neighbors_:
        knn = KNeighborsClassifier(n_neighbors=k)
        scores = cross_val_score(knn, X_train, y_train, 
                                 cv=k_fold, scoring='accuracy')
        cv_scores.append(scores.mean())
    
    # calculate misclassification errors
    errors = [1 - x for x in cv_scores] 
    # determining the best value of k
    optimal_k = neighbors_[errors.index(min(errors))] 
    
    # plot misclassifictin error vs k
    plt.plot(neighbors_, errors)
    plt.xlabel('Number of Neighbors K')
    plt.ylabel('Missclassification error')
    plt.show()

    return optimal_k

def plot_iris_data(X, y, idx_1, idx_2, col_1, col_2, taget_names):
    # ------------------- #
    # PLOT TRAINING DATA 2D
    # ------------------- #
    colormap = np.array(['red', 'green', 'blue'])
    
    red_patch = mpatches.Patch(color='red', label=target_names[0])
    green_patch = mpatches.Patch(color='green', label=target_names[1])
    blue_patch = mpatches.Patch(color='blue', label=target_names[2])

    plt.scatter(X[:, idx_1], X[:, idx_2], c=colormap[y])
    plt.title("The Iris Data Set: " + col_1 + ' vs ' + col_2, fontsize=12)

    plt.xlabel(col_1, fontsize=10)
    plt.ylabel(col_2, fontsize=10)
    plt.rcParams["figure.figsize"] = (6,4)

    ax = plt.subplot(111)
    ax.legend(handles=[red_patch, green_patch, blue_patch], 
              title='Species', loc='upper center', 
              bbox_to_anchor=(1.2,1), shadow=True, ncol=1)
    # bbox_to_anchor=(a,b) => put legend outside plot.

    plt.show()
    
# plot data with boundaries
def plot_data_boundaries(X, y, k):
    h = .02 # step size in the mesh
    n_neighbors = k # number of neighbors (k)

    # Create color maps
    cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
    cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])

    for weights in ['uniform', 'distance']:
        # we create an instance of Neighbors Classifier and fit the data
        clf = neighbors.KNeighborsClassifier(n_neighbors, weights=weights)
        clf.fit(X, y)

        # Plot the decision boundary. 
        # For that, we will assign a color to each
        # point in the mesh [x_min, x_max]x[y_min, y_max].
        x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
        y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
        xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                             np.arange(y_min, y_max, h))

        Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])

        # Put the result into a color plot
        Z = Z = Z.reshape(xx.shape)
        plt.figure()
        plt.pcolormesh(xx, yy, Z, cmap=cmap_light)

        # Plot also the training points
        plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold,
                    edgecolor='k', s=20)

        plt.xlim(xx.min(), xx.max())
        plt.ylim(yy.min(), yy.max())

        plt.title("3-Class classification (k = %i, weights = '%s')"
                  % (n_neighbors, weights))

    plt.show()

# =============================
# Implementing kNN from scratch
# =============================
def train(X_train, y_train):
    # do nothing
    return

def predict(X_train, y_train, x_test, k):
    distances = []
    targets = []

    for i in range(len(X_train)):
        # calculate Euclidean distance between x_test and ith 
        # observation, add it into distances list
        distance = np.sqrt(np.sum(np.square(x_test - X_train[i])))
        distances.append([distance, i]) 
        
    # sort the list
    distances = sorted(distances)
    
    # make a list of the k neighbors' targets
    for i in range(k):
        index = distances[i][1]
        targets.append(y_train[index])
        
    # return most common target
    return Counter(targets).most_common(1)[0][0]

def kNearestNeighbor(X_train, y_train, X_test, predictions, k):
    # train the input data
    train(X_train, y_train)
    
    # loop over all test observations
    for i in range(len(X_test)):
        predictions.append(predict(X_train, y_train, X_test[i, :], k))
    
    return predictions
```

```python
# split dataset into training and test set
# Input: iris.data.txt (https://archive.ics.uci.edu/ml/
# machine-learning-databases/iris/iris.data)
# Output: Classification by kNN

# setup k_fold cross validation
k_fold = 10

# get iris data
iris = datasets.load_iris() # load iris data
iris_data = iris.data
iris_target = iris.target
target_names = iris.target_names
feature_names = iris.feature_names

# convert iris_data and iris_target into X, y
# set observations' attributes
X_col_names = ['speal_length', 'sepal_width', 'petal_length', 
               'petal_width'] 
y_col_name = ['specie']
X = np.array(pd.DataFrame(iris_data, columns=X_col_names).iloc[:, 0:4])
y = np.array(pd.DataFrame(iris_target, columns=y_col_name)['specie'])

# split data into train and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, 
                                                    random_state=42)

# get optimal K
K = optimal_K_neighbors(X_train, y_train, k_fold)

# plot iris data
plot_iris_data(X_train, y_train, 0, 1, X_col_names[0], 
               X_col_names[1], target_names)
plot_iris_data(X_train, y_train, 2, 3, X_col_names[2], 
               X_col_names[3], target_names)
plot_data_boundaries(X_train[:, 0:2], y_train, K)

# predict test observations
predictions = []
y_pred = kNearestNeighbor(X_train, y_train, X_test, predictions, K)

# calculate accuracy (manually)
correct = (y_pred == y_test)
accuracy = correct.sum() / correct.size
print('\nThe accuracy of our classifier is %d%%', accuracy*100)
```
