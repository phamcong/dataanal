```python
# need to convert value "?" to NaN. 
# there are at least two ways to do that: 
# 1. add na_values = ['?'] when load DataFrame.
# 2. df['column_name'] = pd.to_numeric(df['column_name]', errors='coerce')
# (The errors='coerce' parameter instructs Pandas to enter a NaN at any field 
# where the conversion fails.)
cns = pd.read_csv('Datasets/census.data', names=cols, na_values=['?'])
```

```python
ordered_ed = ['1st-4th','5th-6th','7th-8th','9th','10th','11th','12th',
              'HS-grad','Preschool','Some-college','Bachelors','Masters',
              'Doctorate']
```

```python
cns['encoded_education'] = cns['education'].astype('category', ordered=True,
                                           categories=ordered_ed).cat.codes
```

```python
cns.head()
```

```python
cns = cns.drop('encoded_ed',axis=1)
```

```python
cns['encoded_education'].unique()
```

```python
ordered_classif = ['<=50K','>50K']
```

```python
cns['encoded_classification'] = cns['classification'].astype('category',
                        ordered=True, categories=ordered_classif).cat.codes
```
