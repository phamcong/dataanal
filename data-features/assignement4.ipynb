```python
# read_html() returns list of DataFrames.
df = pd.read_html('http://www.espn.com/nhl/statistics/player/_/stat/points/sort/points/year/2015/seasontype/2',skiprows=1)[0]
```

```python
df.head(20)
```

```python
# Drop any row that has at least 4 NON-NaNs within it:
df = df.dropna(axis=0, thresh=4)
```

```python
df2.columns = cols
```
