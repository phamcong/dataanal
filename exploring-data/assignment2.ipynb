```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
```

```python
df = pd.read_csv('Datasets/wheat.data')
```

```python
df.head()
```

```python
plt.scatter(x='area', y='perimeter', data=df)
plt.xlabel('area')
plt.ylabel('perimeter')
```

```python
plt.scatter(x='groove', y='asymmetry', data=df, c='red')
plt.xlabel('groove')
plt.ylabel('asymmetry')
```

```python
plt.scatter(x='compactness', y='width', data=df, c='green')
plt.xlabel('compactness')
plt.ylabel('width')
```
